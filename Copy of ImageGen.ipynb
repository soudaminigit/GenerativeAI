```python
import os

#openai.api_key = os.getenv("OPENAI_API_KEY")
response = openai.Image.create(
  prompt="superman swimming",
  n=1,
  size="256x256"
)
```

```python
url=response.data[0]['url']
url
```

```python
import urllib.request
from PIL import Image
urllib.request.urlretrieve(url, "penguin_image.jpg")

img = Image.open("penguin_image.jpg")

img.show()
```

```python
import os
import openai
response = openai.Image.create_variation(
  image=open("penguin_image.jpg", "rb"),
  n=1,
  size="256x256"
)
```

```python
url=response.data[0]['url']
urllib.request.urlretrieve(url, "penguin_image_alter.jpg")

img = Image.open("penguin_image_alter.jpg")

img.show()
```
